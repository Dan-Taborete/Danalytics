```python
# Exploring data type of movie_release_date

df_movie_list['movie_release_date'].info
```

```python
# Converting movie_release_date from object dtype to datetime

df_movie_list['movie_release_date'] = pd.to_datetime(df_movie_list['movie_release_date'])
```

```python
df_movie_list['movie_release_date']
```

```python
# The first date in the dataset

df_movie_list['movie_release_date'].min()
```

```python
# Creating a seperate column for the movie release year

df_movie_list['movie release year']= df_movie_list['movie_release_date'].dt.year
```

```python
df_movie_list['movie release year']
```

```python
df_movie_list.columns
```

```python
# Seperating and Creating new DataFrame that includes movie information only

df_movie_info = df_movie_list[['movie_id', 'movie_title', 'movie_popularity', 'movie_release_date',
       'movie_budget', 'movie_revenue','movie release year']]
```

```python
df_movie_info
```

```python
# Dropping the duplicates

df_movie_info.drop_duplicates(inplace= True)
```

```python
df_movie_info
```

```python
# Identifying the top 10 based on the movie_revenue

df_movie_info.nlargest(10, 'movie_revenue')
```

```python
# Sorting the values

df_movie_info.sort_values('movie_revenue', inplace=True)
```

```python
df_movie_info
```

```python
# Identifying the highest revenue each year

df_movie_info.groupby('movie release year').last()
```

```python
# Identifying the unique movies per year

df_movie_info.groupby('movie release year')['movie_id'].nunique()
```

```python
# Identifying total revenue per year

top_movie_per_year = df_movie_info.groupby('movie release year')['movie_revenue'].sum()
```

```python
top_movie_per_year
```

```python
# Identifying the top 10 highest revenue of year

top_movie_per_year.nlargest(10)
```

```python
# Identifying the average revenue per year

df_movie_info.groupby('movie release year')['movie_revenue'].mean()
```

```python
# Creating a histogram

df_movie_info['movie_budget'].hist()
```

```python
# Exploring the relation between movie_budget and movie_revenue using scatter plot

sns.relplot(data=df_movie_info, x= 'movie_budget',y= 'movie_revenue')
```

```python
# Exploring the relation between movie_budget and movie_revenue using joint plot


sns.jointplot(data=df_movie_info, x= 'movie_budget',y= 'movie_revenue')
```
